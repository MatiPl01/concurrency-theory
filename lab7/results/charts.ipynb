# Wykresy

```python
from matplotlib import pyplot as plt
import seaborn as sns

def create_chart(X, Y, label, xlabel, ylabel, title, color='b', ax=plt):
    ax.plot(X, Y, color, label=label)

    if label: 
        ax.legend(loc='best')
        
    if ax is plt:
        ax.title(title)
        ax.xlabel(xlabel)
        ax.ylabel(ylabel)
    else:
        ax.title.set_text(title)
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)

    ax.grid()
    sns.despine()
    
def create_charts(data, x_key, y_key, xlabel, ylabel, title, colors, ax=plt):
    for (key, values), color in zip(data.items(), colors):
        label = f"bufor {key.split('-')[1]}-elementowy"
        create_chart(values[x_key], values[y_key], label, xlabel, ylabel, title, color, ax)
        
def comparison_chart(data_list, x_key, y_key, xlabel, ylabel, title, colors, ax=plt):
    for data, color in zip(data_list, colors):
        values = data['data']
        label = data['label']
        create_chart(values[x_key], values[y_key], label, xlabel, ylabel, title, color, ax)
    ax.grid()
    sns.despine()
```
